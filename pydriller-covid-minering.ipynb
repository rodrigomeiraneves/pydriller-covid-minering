# Mining software repositories


```python
def fetch_detached_repositories(termo='covid', limit=10):
    query = f"{termo} in:name,description"
    repos = g.search_repositories(query=query, sort="stars", order="desc")
    return repos[:limit]

def mining_info_repository(repo):
    print(f"Name of projeto: {repo.name}")
    print(f"URL: {repo.html_url}")
    print(f"Description: {repo.description}")

    linguagens = repo.get_languages()
    print(f"Programming languages: {', '.join(linguagens.keys()) if linguagens else 'Not listed'}")

    technologies = repo.get_topics()
    print(f"Technologies involved: {', '.join(technologies) if technologies else 'Not listed'}")

    collaborators = repo.get_contributors().totalCount
    print(f"Number of collaborators: {collaborators}")

    commits = repo.get_commits().totalCount
    print(f"Number of commits: {commits}")

    last_update = repo.updated_at
    print(f"Last update: {last_update.strftime('%Y-%m-%d %H:%M:%S')}")

    print(f"Number of stars: {repo.stargazers_count}")
    print(f"Number of watchers: {repo.subscribers_count}")
    print(f"Number of forks: {repo.forks_count}")
    print(f"Number of branches: {repo.get_branches().totalCount}")

    print("-" * 80)

highlighted_covid_repositories = fetch_detached_repositories()

for repo in highlighted_covid_repositories:
    mining_info_repository(repo)
```
